## KDE (Kernel Density Estimator) Plots

KED plots are mostly used with histogram which can expression multidimensional data with only one dimension. 
- Much easier to observe the distribution

## Histogram

```python

import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns

set_one = np.random.normal(0, 1, 100)
set_two = np.random.normal(0, 1, 100)
set_three = np.random.exponential(1, 100)


sns.kdeplot(set_one, shade = True)
sns.kdeplot(set_two, shade = True)
sns.kdeplot(set_three, shade = True)

plt.legend(['set_one', 'set_two', 'set_three'])

plt.show()

```

## Box Plot

```python
n = 100
df = pd.DataFrame({
    'label':['set_one'] * n + ['set_two'] * n + ['set_three'] * n,
    'value':np.concatenate([set_one, set_two, set_three])
    
})

sns.boxplot(data=df, x = 'label', y = 'value')
```

## Voilin Plots

- Voilin Plot is essentiall **two KDE(histogram) plots symmetric** by the center black line
- The center **white dot** is where the **median** lies
- the **thick black line** is the **interquartile** range
- the **black line** extends from the white dot in the center is the (by default) **95% confidence interval**

```python
sns.violinplot(data=df, x = 'label', y = 'value')
plt.show()
```
