```python
import pandas as pd
import seaborn as sns

df = pd.read_csv("https://raw.githubusercontent.com/Codecademy/datasets/master/streeteasy/manhattan.csv")
df.head()
```

```python
print(df.dtypes)
df['bedrooms'].value_counts()
```

```python
df.borough.unique()
```

```python
from sklearn.model_selection import train_test_split

x = df[['bedrooms', 'bathrooms', 'size_sqft', 'min_to_subway', 'floor', 'building_age_yrs', 'no_fee', 'has_roofdeck', 'has_washer_dryer', 'has_doorman', 'has_elevator', 'has_dishwasher', 'has_patio', 'has_gym']]
y = df['rent']

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, test_size=0.2)

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

mlr = LinearRegression()
mlr.fit(x_train, y_train)

y_predict = mlr.predict(x_test)

```

```python
from matplotlib import pyplot as plt



plt.scatter(y_test, y_predict, alpha = 0.5)

plt.xlabel('actual price')

plt.ylabel('predicted price')

plt.show()


```

```python
print(mlr.coef_)
```

```python
plt.scatter(df[['size_sqft']], df[['rent']], alpha=0.4)
```

```python
plt.scatter(df[['min_to_subway']], df[['rent']], alpha=0.4)
```

```python
print("Train score:")
print(mlr.score(x_train, y_train))
```

```python
print("Train score:")
print(mlr.score(x_test, y_test))

SSR = ((y_test - y_predict) ** 2).sum()

SST = ((y_test - y_test.mean()) ** 2).sum()

print(1- SSR/SST)
```

```python
residuals = y_predict - y_test

plt.scatter(y_predict, residuals, alpha = 0.4)
```

```python
import pandas as pd
import seaborn as sns

df = pd.read_csv("https://raw.githubusercontent.com/Codecademy/datasets/master/streeteasy/manhattan.csv")
df.head()

from sklearn.model_selection import train_test_split

x = df[['bedrooms', 'bathrooms', 'size_sqft', 'floor']]
y = df['rent']

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, test_size=0.2)

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)

from sklearn.linear_model import LinearRegression

mlr = LinearRegression()
mlr.fit(x_train, y_train)

y_predict = mlr.predict(x_test)


mlr.score(x_test, y_test)
```
